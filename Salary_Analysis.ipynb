```python
# Removing Missing Values Of Salary Estimates (-1, if salary is not given insted of NaN)
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

Okhay! We have only 1 missing value. Let's move on!!

Now we have to remove `$` , `K` & `(Glassdoor est.)`. Here I have created a general concept to do this.

```python
# string = '$200K-$520K (Glassdoor est.)'
# new_str = string.split('(')[0].replace('$', '').replace('K', '').lower().replace('employer provided salary', '').replace('per hour', '').split('-')
# lowerValue, upperValue = int(new_str[0]), int(new_str[1])
# averageSalary = (lowerValue + upperValue)/2
# print(averageSalary)
```

```python
def salaryCleaner(object):
    string1 = object.split('(')[0].replace('$', '').replace('K', '').split('-')
    averageSalary = (int(string1[0]) + int(string1[1]))/2
    return averageSalary


salaryCleaner('200-220 (Glassdoor est.)')
```

```python
df['Average Salary'] = df['Salary Estimate'].apply(lambda x: salaryCleaner(x))
```

```python
df.head(120)
```

So we cleaned `Salary Estimate`. Next we'll clean `Company Name`. We need to extract only company name from this column. Let's do it

```python
df3 = df2.copy()
```

```python
def jobTitleCleaner(object):
    job = object
    if 'data' in job.lower()and 'analyst' in job.lower():
        return 'Data Analyst'

def jobPositionCleaner(object):
    job = object
    if 'senior' in job.lower():
        return 'Senior'
    elif 'junior' in job.lower():
        return 'Junior'
    elif 'entry level' in job.lower() or 'intern' in job.lower():
        return 'Intern'
    else:
        return 'Na'

jobPositionCleaner('Data Business analyst')
```

```python
df3['Profession'] = df3['Job Title'].apply(lambda x: 'Data Analyst' if 'data' and 'analyst' in x.lower() else x)
```

```python
df3['Profession'].value_counts()
```

```python
df3['Job Position'] = df3['Job Title'].apply(jobPositionCleaner)
```

```python
df3
```

```python
df3['Job Position'].value_counts()
```

```python
df3['Job State'] = df3['Job State'].apply(lambda x: x.strip() if x.strip().lower() != 'arapahoe' else 'CO')
```

```python
df3['Job State'].value_counts()
```

```python
def competitorsCount(object):
    if object == '-1':
        return 0
    else:
        x = object.split(',')
        count = len(x)
        return int(count)

competitorsCount('HUL, Dabur')
```

```python
df3['Competitors Count'] = df3['Competitors'].apply(competitorsCount)
```
